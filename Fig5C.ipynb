```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

data = pd.read_csv('Fig5C.csv')
data.head()

# Set up the plot
plt.figure(figsize=(8, 6))

# Plot a jointplot with regression line
sns.jointplot(
    data=data, 
    x='Average_Repellency_Anopheles', 
    y='Average_Repellency_ATC', 
    kind='reg', 
    height=8, 
    scatter_kws={'s': 60, 'alpha': 0.7},
    line_kws={"color": "red", "linewidth": 1}
)

plt.axhline(y=50, color='red', linestyle='--')
plt.axvline(x=70, color='red', linestyle='--')

# Add main title
plt.suptitle('Joint Plot of Average Repellency for Anopheles vs ATC', y=1.02)
plt.show()
```
